```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.power(2,3)
```

```python
x = np.array([1,2,3,4,5,6])
y = np.power(x,3)
plt.plot(x,y)
plt.show()
```

```python
# Sin curve
x = np.linspace(0,10,1000)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
# Cos curve
x = np.linspace(0,10,100)
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

```python

x = np.linspace(0,10,1000)
y = np.sin(x)
plt.plot(x,y)
plt.show()

x = np.linspace(0,10,100)
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

```python

x = np.linspace(0,10,1000)
y = np.cos(x)
z = np.sin(x)
plt.plot(x,y,z)
plt.show()
```

```python

x = np.linspace(0,10,1000)
plt.figure(figsize=(15,6))
y = np.cos(x)
z = np.sin(x)
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```
